```python
import datarobot as dr 
import pandas as pd
import numpy as np
import math
```

```python
feature_width = 2000
np.random.seed(0)  # 做實驗這個很重要！固定 dataset.
x = np.random.rand(300,feature_width)
data = pd.DataFrame(np.array(x))
data.head()
```

# 大於小於 0.5 

不能更簡單了，純亂數 2000 features 其中一條以 0.5 為界，大於等於 0.5 者 target 為 True.

```python
# 切一半，是否 >= 0.5 
for i in range(data.shape[0]):
    data.loc[i,0] = True if data.loc[i,1] >= 0.5 else False

data.head()
```

```python
# prediction 結果跟 dataset 比對。用 prediction 0 的 index 去算 dataset 的 sum 就是 FN
indexs = list(df_predict_holdout.query('prediction == 0.0')['row_id'])
%f indexs count nip -->
positives = [data[target_name].iloc[i] for i in indexs]    
%f # FN 越少越好
sum(positives)
```

```python
# prediction 結果跟 dataset 比對。用 prediction 1 的 index 去算 dataset 的 sum 就是 TP
indexs = list(df_predict_holdout.query('prediction == 1.0')['row_id'])
%f indexs count nip -->
positives = [data[target_name].iloc[i] for i in indexs]    
%f # TP 越接近越好
sum(positives)
```

```python
# model_id 用手工上 UI 取得
model_id = '60d5730ea597d43c877de02a'
model = dr.Model.get(project=project.id,model_id=model_id)
model
```

```python
training_predictions_job = model.request_training_predictions(dr.enums.DATA_SUBSET.HOLDOUT)
training_predictions = training_predictions_job.get_result_when_complete()
training_predictions
```

```python
# Fetch training predictions as data frame
df_predict_holdout = training_predictions.get_all_as_dataframe()
df_predict_holdout.head()
```

```python
# 照 holdout 的 index 到 data 裡去把標準答案 target 挑出來
indexs = list(df_predict_holdout['row_id'])
%f indexs count nip -->
targets = [data[2].iloc[i] for i in indexs]    
targets[:4]
```

```python
# 包裝成 DataFrame 
answers = pd.DataFrame(targets, index=indexs, columns=['target'])
answers.head()
```

```python
# 並放目視比對
results = pd.merge(df_predict_holdout,answers, left_on='row_id', right_index=True)
results.head()
```

```python
%f results :> ['prediction'][0] -->
%f results :> ['target'][0] -->
    
```

```python
# 弄出新的 column 'is_bigger' 
# https://stackoverflow.com/questions/51499385/how-to-add-values-to-a-new-column-in-pandas-dataframe
results.at[0, 'is_bigger'] = True
# df.at[index, 'new_column'] = new_value
```

```python
results.head()
```

```python
for i in range(results.shape[0]):
    results.loc[i,'is_bigger'] = results['prediction'][i] > results['target'][i]
    # data.loc[i,2] = math.sqrt(data.loc[i,3])
```
