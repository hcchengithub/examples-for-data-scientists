```python
import pandas as pd
import numpy as np
import math
```

```python
raw_data = pd.read_excel('Copy of Result_Rosa_NB -20210609 Tool.xlsx')
raw_data
```

```python
raw_targets = pd.read_excel('Copy of Result_Rosa_NB -20210609 Tool.xlsx',sheet_name=1)
raw_targets
```

```python
devices = set([x for x in raw_targets['P2']])
%f devices -->
```

```python
%f devices count nip -->
```

```python
try:
    devices.remove(np.nan)
except:
    pass
```

```python
%f devices -->
```

```python
%f devices count nip -->
```

```python
devices = tuple(devices)  # 固定順序
```

```python
%f devices -->
```

```python
# 拆分 P2 成個別 device 的 list 準備作為 target column 
classes = []
for dd in raw_data['P2(What_Device)']:
    row = []
    for i in range(len(devices)):
        if type(dd) == str and devices[i] in dd:
            row.append(1)
        else:
            row.append(0)
    classes.append(row)
%f classes count nip -->
%f raw_data count nip -->
```

```python
%f classes :> [7] -->
```

```python
targets = pd.DataFrame(np.array(classes), columns = devices)
targets
```

```python
data = pd.merge(raw_data['Title'],targets, left_index=True, right_index=True)
data 
```

```python
# 對照檢查無誤
data.to_excel("1.xls")
```
