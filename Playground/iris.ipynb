```python
import datarobot as dr 
import pandas as pd
import time, io
```

```python
df = pd.read_csv('iris-data.csv')  # also iris-holdout.csv
df
```

```python
dr.Client(token=peforth.execute('api_token').pop(), endpoint=peforth.execute('endpoint').pop())
```

```python
# Create project 一定要給 dataset, 這是 DataRobot 的觀念， dataset 就是 project 的靈魂。
# 這一步過了好久，上網頁查看，發現 DataRobot_APITraining 已經建立起來了，再過一會兒發現這裡也好了。
# [x] 重複做會重複 create 同名 project 所以要研究如何開舊的？ <--- get project id 即可。
project = dr.Project.create(df, project_name = 'iris') # 第一次用 create() 的，以後 get project id 即可。
```

```python
# 再來就是 Start() 了，set_target() 就是 start 簡捷俐落。
project.set_target(  
    target='class',
    worker_count = '-1', 
    mode=dr.AUTOPILOT_MODE.QUICK
)
```

```python
# check if jobs are all done?
jobs = project.get_all_jobs()
%f jobs -->
%f jobs count nip --> # 檢查這個長度就知道 project 是否還在 busy

# check if jobs are all done? 這個辦法不太靈，modeling 會全部結束，然後又開始，這個 loop 就會被甩出來。可能需要兩層？
while len(project.get_all_jobs()) > 0:
    time.sleep(10)
```
